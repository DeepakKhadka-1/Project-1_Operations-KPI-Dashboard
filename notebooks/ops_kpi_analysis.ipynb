```python
## 1. IMPORT LIBRARIES
import pandas as pd
import numpy as np

```

```python
##2. LOAD DATA
file_path = r"C:\Users\Deepak khadka\Desktop\DA INTERVIEW PORTOFOLIO\Projects\Project-1_Business-Ops-KPI\data\raw\Retail-Supply-Chain-Sales-Dataset.xlsx"

df = pd.read_excel(file_path)
```

```python
##3. BASIC CHECKS
print("Shape:", df.shape)
df.head()
df.info()
df.isna().sum()



```

```python
##4. STANDARDISE COLUMN NAMES
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("-", "_")
)

list(df.columns)
```

```python
##5. DATE COLUMNS
df[["order_date", "ship_date"]].dtypes
```

```python
##6. CREATE OPERATIONS METRICS

#6.1 Delivery Time (Days)
df["delivery_days"] = (df["ship_date"] - df["order_date"]).dt.days

#6.2 SLA Definition
df["sla_days"] = 7

#6.3 Delay Calculation
df["delay_days"] = df["delivery_days"] - df["sla_days"]

#6.4 Delivery Status
df["delivery_status"] = np.where(df["delay_days"] <= 0, "On-Time", "Delayed")

```

```python
## 7. KPI CALCULATIONS

#7.1 SLA Compliance %
sla_compliance = (
    df["delivery_status"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
)

sla_compliance

#7.2 Average Delivery Time
avg_delivery_time = df["delivery_days"].mean().round(2)
avg_delivery_time

#7.3 Delay Rate by Region
delay_by_region = (
    df.groupby("region")["delivery_status"]
      .apply(lambda x: (x == "Delayed").mean() * 100)
      .round(2)
      .reset_index(name="delay_percentage")
)

delay_by_region


#7.4 Delay Rate by Category
delay_by_category = (
    df.groupby("category")["delivery_status"]
      .apply(lambda x: (x == "Delayed").mean() * 100)
      .round(2)
      .reset_index(name="delay_percentage")
)

delay_by_category


```

```python
##8. SAVE CLEAN DATA
processed_path = r"C:\Users\Deepak khadka\Desktop\DA INTERVIEW PORTOFOLIO\Projects\Project-1_Business-Ops-KPI\data\processed\cleaned_operations_data.csv"

df.to_csv(processed_path, index=False)

processed_path
```
